<font size = 6> Out of Bag Error</font>

```python
best_rf = load_pickle('trained_models\\rf_best.pkl')
```

```python
to = load_pickle('..\\prepared_data\\to.pkl')
```

```python
def r_mse(pred,y): # I will be using root mean square error to evaluate the perform of my model
    return round(math.sqrt(((pred-y)**2).mean()), 6)
def model_rmse(model, xs, y): 
    return r_mse(model.predict(xs), y)
```

```python
model_rmse(best_rf, to.valid.xs, to.valid.y)
```

```python
# if OOB error is lower than validation set error, something else may be causing our validation set error 
# than just normal generalization difficulty 

r_mse(best_rf.oob_prediction_, to.train.y)
```

It would appear that our validation set error is due to general challenges with generalization

<font size = 6> Feature Importance </font>

```python
best_rf.feature_importances_
```

```python
# that's hard to read, so fast.ai does
def rf_feat_importance(m, df):
    return pd.DataFrame({'cols':df.columns, 'imp':m.feature_importances_}
                       ).sort_values('imp', ascending=False)
```

```python
fi = rf_feat_importance(best_rf, to.valid.xs)
fi
```

```python
def plot_fi(fi):
    return fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)

plot_fi(fi[:30]);
```
