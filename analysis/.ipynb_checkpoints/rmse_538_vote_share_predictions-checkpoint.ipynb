```python
import pandas as pd
import numpy as np
```

```python
# FiveThirtyEight's predictions for the vote share of each candidate in the 2020 US Senate elections
preds_538 = pd.read_csv('..\\data\\senate_state_toplines_2020.csv')
preds_538
```

```python
# reducing to needed columns and to only the predictions made right before election day
preds_538 = preds_538.loc[:34,['name_D1','name_R1','voteshare_mean_D1','voteshare_mean_R1']]
preds_538
```

```python
preds_538.drop(index=32, inplace=True) # drop AR election since Tom Cotton didn't have an opponent
preds_538
```

Loading actual 2020 senate results, also from FiveThirtyEight so that consistent data format to reduce data preparation time

```python
actual = pd.read_csv('..\\data\\actual2020_election_results_senate.csv')
actual
```

```python
def r_mse(pred,y): 
    return round(np.sqrt(((pred-y)**2).mean()), 6)
```

```python
def extract_last_name(name: str) -> str:
    splt = name.split(' ')
    if splt[-1] in ['Jr.','Sr.']:
        final = splt[-2].replace(',', '') # in case comma after last name before JR./III
        return final
    else:
        return splt[-1]
```

```python
def get_share(name):
    last_name = extract_last_name(name)
    first_name = name.split(' ')[0]
    matches = actual[(actual.cycle==2020) & (actual.candidate_name.str.contains(last_name,regex=True,na=False)) & (actual.candidate_name.str.contains(first_name,regex=True,na=False))]
    if len(matches)!=0:
        share = matches.iloc[0,16]
        return share
    return 50
```

```python
preds_538['actual_dem_share'] = preds_538.apply(lambda x: get_share(x.name_D1), axis=1)
preds_538['actual_repub_share'] = preds_538.apply(lambda x: get_share(x.name_R1), axis=1)
```

```python
preds_538
```

```python
# manually specifying the unfound names
preds_538.loc[3,'actual_dem_share'] = 43.86 # MJ Hegar Texas
preds_538.loc[5,'actual_repub_share'] = 65.73 # Mike Rounds South Dakota
preds_538.loc[9,'actual_repub_share'] = 62.91 # Jim Inhofe
preds_538.loc[25,'actual_repub_share'] = 62.62 # Jim Risch
```

```python
# manually correcting one where there were two Kevin O'Connors, just diff middle initials
preds_538.loc[20, 'actual_repub_share'] = 33.04
```

```python
preds_538['predicted_repub_two_party'] = preds_538['voteshare_mean_R1']/(preds_538['voteshare_mean_R1']+preds_538['voteshare_mean_D1'])
preds_538['actual_repub_two_party'] = preds_538['actual_repub_share']/(preds_538['actual_repub_share']+preds_538['actual_dem_share'])
```
